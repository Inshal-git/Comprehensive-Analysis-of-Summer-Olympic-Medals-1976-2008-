```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV to Excel
csv_file_path = 'C:/Projects/Summer-Olympic-medals-1976-to-2008.csv'
excel_file_path = 'C:/Projects/Summer-Olympic-medals-1976-to-2008.xlsx'
```

```python
data = pd.read_csv(csv_file_path, encoding='latin1')
```

```python
data.to_excel(excel_file_path, index=False)
```

```python
df = pd.read_excel(excel_file_path)
```

```python
print(df.head())
```

```python
medal_count = df['Country'].value_counts()
print(medal_count)
```

```python
top_10_countries = medal_count.head(10)

plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.title('Top 10 Countries by Medal Count (1976-2008)')
plt.show()
```

```python
gender_medal_count = df['Gender'].value_counts()

plt.figure(figsize=(6, 6))
gender_medal_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Medal Distribution by Gender (1976-2008)')
plt.ylabel('')
plt.show()
```

```python
top_athletes = df['Athlete'].value_counts().head(10)

plt.figure(figsize=(12, 6))
top_athletes.plot(kind='bar')
plt.xlabel('Athlete')
plt.ylabel('Number of Medals')
plt.title('Top 10 Athletes by Medal Count (1976-2008)')
plt.show()
```

```python
medal_count_by_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
medal_count_by_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.title('Total Medal Count Over the Years (1976-2008)')
plt.show()
```

```python
medal_count_by_sport = df['Sport'].value_counts().head(10)

plt.figure(figsize=(12, 6))
medal_count_by_sport.plot(kind='bar')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.title('Top 10 Sports by Medal Count (1976-2008)')
plt.show()
```

```python
countries = ['United States', 'China', 'Russia']
df_specific_countries = df[df['Country'].isin(countries)]

for country in countries:
    df_country = df_specific_countries[df_specific_countries['Country'] == country]
    medal_count_by_year_country = df_country['Year'].value_counts().sort_index()

    plt.figure(figsize=(12, 6))
    medal_count_by_year_country.plot(kind='bar')
    plt.xlabel('Year')
    plt.ylabel('Number of Medals')
    plt.title(f'Medal Count Over the Years for {country} (1976-2008)')
    plt.show()
```

```python
event_gender_medal_count = df['Event_gender'].value_counts()

plt.figure(figsize=(6, 6))
event_gender_medal_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Medal Distribution by Event Gender (1976-2008)')
plt.ylabel('')
plt.show()
```

```python
medal_type_count = df['Medal'].value_counts()

plt.figure(figsize=(6, 6))
medal_type_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#FFD700', '#C0C0C0', '#CD7F32'])
plt.title('Medal Distribution by Medal Type (1976-2008)')
plt.ylabel('')
plt.show()
```

```python
top_sports = df['Sport'].value_counts().head(3).index
plt.figure(figsize=(12, 6))
for sport in top_sports:
    df_sport = df[df['Sport'] == sport]
    medal_count_by_year_sport = df_sport['Year'].value_counts().sort_index()
    plt.plot(medal_count_by_year_sport, marker='o', label=sport)

plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.title('Medal Trends for Top 3 Sports Over the Years (1976-2008)')
plt.legend()
plt.show()
```
